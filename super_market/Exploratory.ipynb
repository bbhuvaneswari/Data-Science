```python
dw_mapping={
    0:"Mon",
    1:"Tue",
    2:"Wed",
    3:"Thur",
    4:"Fir",
    5:"Sat",
    6:"Sun"
}
```

```python
data["Day_of_week"]=data["Date"].dt.dayofweek.map(dw_mapping)
```

```python
data["Day_of_week"].value_counts().plot(kind="bar")
```

```python
month_mapping={
    1:"Jan",
    2:"Feb",
    3:"March"
}
```

```python
data["month"]=data["Date"].dt.dayofweek.map(month_mapping)
```

```python
data["month"].value_counts().plot(kind="bar")
```
